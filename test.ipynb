```python
import time
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC


# Specify the path to your Chrome user profile
chrome_options = webdriver.ChromeOptions()
chrome_options.add_argument(r"--user-data-dir=C:\Users\Humza\AppData\Local\Google\Chrome\User Data")
chrome_options.add_argument(r"--profile-directory=Profile 1")  # Or the specific profile directory

# Initialize WebDriver with Chrome profile
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=chrome_options)

# Search query
SEARCH_QUERY = "software engineer"

# Navigate to LinkedIn
driver.get("https://www.linkedin.com")

# Helper: Perform search and navigate to people tab
def search_people(query):
    search_box = driver.find_element(By.CSS_SELECTOR, "#global-nav-typeahead > input")
    search_box.send_keys(query)
    search_box.send_keys(Keys.ENTER)
    time.sleep(5)
    people_tab = driver.find_element(By.CSS_SELECTOR, "#search-reusables__filters-bar > ul > li:nth-child(1) > button")
    people_tab.click()
    time.sleep(5)

# Helper: Scrape names from the current page
def scrape_names():
    names = []
    results = driver.find_elements(By.CSS_SELECTOR, ".entity-result__title-text a span span")
    for result in results:
        names.append(result.text)
    return names

# Helper: Navigate to the next page
def go_to_next_page():
    try:
        next_button = WebDriverWait(driver, 10).until(
            EC.element_to_be_clickable((By.XPATH, "//button[@aria-label='Next']"))
        )
        next_button.click()
        time.sleep(5)  # Allow some time for the next page to load
        return True
    except Exception as e:
        print("Next page not found or end of results:", e)
        return False

# Main logic
def main():
    search_people(SEARCH_QUERY)
    time.sleep(10)

    all_names = []
    page = 1

    while page <= 100:  # Limit to 100 pages
        print(f"Scraping page {page}...")
        names = scrape_names()
        all_names.extend(names)

        # Break if no next page
        if not go_to_next_page():
            break

        page += 1

    # Save results to CSV
    df = pd.DataFrame(all_names, columns=["Name"])
    df.to_csv("linkedin_names.csv", index=False)
    print("Scraping complete. Results saved to 'linkedin_names.csv'.")

# Run the script
try:
    main()
finally:
    driver.quit()
```
