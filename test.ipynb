```python
import csv
import time
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

# Specify the path to your Chrome user profile
chrome_options = webdriver.ChromeOptions()
chrome_options.add_argument(r"--user-data-dir=C:\Users\Humza\AppData\Local\Google\Chrome\User Data")
chrome_options.add_argument(r"--profile-directory=Profile 1")  # Or the specific profile directory

# Initialize WebDriver with Chrome profile
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=chrome_options)

# Output CSV file
output_file = "linkedin_profiles1.csv"

# Initialize CSV writer
with open(output_file, mode="w", newline="", encoding="utf-8") as csvfile:
    fieldnames = ["Name", "URL", "Experience"]
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
    writer.writeheader()  # Write header once

# Search query
SEARCH_QUERY = "software engineer"

# Navigate to LinkedIn
driver.get("https://www.linkedin.com")
time.sleep(5)  # Allow time for the page to load

# Helper: Perform search and navigate to people tab
def search_people(query):
    search_box = driver.find_element(By.CSS_SELECTOR, "#global-nav-typeahead > input")
    search_box.send_keys(query)
    search_box.send_keys(Keys.ENTER)
    time.sleep(5)
    people_tab = driver.find_element(By.CSS_SELECTOR, "#search-reusables__filters-bar > ul > li:nth-child(1) > button")
    people_tab.click()
    time.sleep(5)

# Helper: Scroll down to load more profiles on the current page
def scroll_down():
    try:
        # Scroll to the bottom of the page
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        time.sleep(3)  # Wait for content to load
        driver.execute_script("window.scrollTo(0, 0);")
        time.sleep(2)
    except Exception as e:
        print(f"Error while scrolling: {e}")

# Helper: Open profiles one by one and save data
def open_profiles():
    try:
        # Locate parent divs with class "mb1"
        profile_containers = driver.find_elements(By.CSS_SELECTOR, "div.mb1")
        unique_urls = set()

        for container in profile_containers:
            try:
                profile_link = container.find_element(By.CSS_SELECTOR, "a.EvQUJBaxIRgFetdTQjAXvpGhCNvVbYEbE")
                profile_url = profile_link.get_attribute("href")

                if profile_url and "/in/" in profile_url and profile_url not in unique_urls:
                    unique_urls.add(profile_url)  # Add to the set

                    # Open the profile in a new tab
                    driver.execute_script(f"window.open('{profile_url}', '_blank');")
                    time.sleep(2)

                    # Switch to the new tab
                    driver.switch_to.window(driver.window_handles[-1])
                    time.sleep(5)

                    # Scrape the name and experience
                    name, experiences = "", []

                    try:
                        main_div = driver.find_element(By.CSS_SELECTOR, "div.ph5.pb5")
                        name_element = main_div.find_element(By.CSS_SELECTOR, "h1.lSVmRsHWVyNnNULXTOLIQEQVzhkaazrTFzDYEg")
                        name = name_element.text
                    except Exception as e:
                        print(f"Error scraping name: {e}")

                    try:
                        experience_divs = driver.find_elements(By.CSS_SELECTOR, "div.display-flex.flex-column.align-self-center.flex-grow-1")
                        for experience in experience_divs:
                            try:
                                title_element = experience.find_element(By.CSS_SELECTOR, "div.display-flex.align-items-center.mr1.hoverable-link-text.t-bold span[aria-hidden='true']")
                                title = title_element.text
                                experiences.append(title)
                            except Exception as e:
                                print(f"Error scraping experience title: {e}")
                    except Exception as e:
                        print(f"Error accessing experience section: {e}")

                    # Write to CSV
                    with open(output_file, mode="a", newline="", encoding="utf-8") as csvfile:
                        writer = csv.DictWriter(csvfile, fieldnames=["Name", "URL", "Experience"])
                        writer.writerow({
                            "Name": name,
                            "URL": profile_url,
                            "Experience": "; ".join(experiences)
                        })

                    # Close the current tab
                    driver.close()
                    driver.switch_to.window(driver.window_handles[0])
                    time.sleep(2)
            except Exception as e:
                print(f"Error processing profile: {e}")
    except Exception as e:
        print(f"Error while opening profiles: {e}")

# Helper: Navigate to the next page
def go_to_next_page():
    try:
        next_button = WebDriverWait(driver, 10).until(
            EC.element_to_be_clickable((By.CSS_SELECTOR, ".artdeco-pagination__button--next"))
        )
        next_button.click()
        time.sleep(5)  # Allow some time for the next page to load
        return True
    except Exception as e:
        print("Next page not found or end of results:", e)
        return False

# Main logic
def main():
    search_people(SEARCH_QUERY)
    time.sleep(10)

    page = 1

    while page <= 100:  # Limit to 100 pages
        print(f"Processing page {page}...")
        scroll_down()  # Scroll down to ensure all profiles are loaded
        open_profiles()

        if not go_to_next_page():
            break

        page += 1

    print("Profile visiting complete.")

# Run the script
try:
    main()
finally:
    driver.quit()
```

```python
import time
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

# Specify the path to your Chrome user profile
chrome_options = webdriver.ChromeOptions()
chrome_options.add_argument(r"--user-data-dir=C:\Users\Humza\AppData\Local\Google\Chrome\User Data")
chrome_options.add_argument(r"--profile-directory=Profile 1")  # Or the specific profile directory

# Initialize WebDriver with Chrome profile
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=chrome_options)

# Search query
SEARCH_QUERY = "software engineer"

# Navigate to LinkedIn
driver.get("https://www.linkedin.com")
time.sleep(5)  # Allow time for the page to load

# Helper: Perform search and navigate to people tab
def search_people(query):
    search_box = driver.find_element(By.CSS_SELECTOR, "#global-nav-typeahead > input")
    search_box.send_keys(query)
    search_box.send_keys(Keys.ENTER)
    time.sleep(5)
    people_tab = driver.find_element(By.CSS_SELECTOR, "#search-reusables__filters-bar > ul > li:nth-child(1) > button")
    people_tab.click()
    time.sleep(5)

# Helper: Scroll down to load more profiles on the current page
def scroll_down():
    try:
        # Scroll to the bottom of the page
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        time.sleep(3)  # Wait for content to load
        # Scroll back to the top in case additional elements load dynamically
        driver.execute_script("window.scrollTo(0, 0);")
        time.sleep(2)
    except Exception as e:
        print(f"Error while scrolling: {e}")

# Helper: Scroll to ensure experience section is loaded
def scroll_to_load_experience():
    try:
        # Scroll to the bottom to make sure all dynamic content (like experience section) is loaded
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        time.sleep(3)  # Wait for content to load
        # You may also scroll back to the top in some cases
        driver.execute_script("window.scrollTo(0, 0);")
        time.sleep(2)
    except Exception as e:
        print(f"Error while scrolling to load experience section: {e}")

# Updated Helper: Open profiles one by one and close tabs after processing
def open_profiles():
    try:
        # Locate parent divs with class "mb1"
        profile_containers = driver.find_elements(By.CSS_SELECTOR, "div.mb1")
        print(f"Found {len(profile_containers)} profile containers on the page.")  # Debugging: Number of containers found

        # Use a set to track unique profile URLs
        unique_urls = set()

        for container in profile_containers:
            try:
                # Navigate to the specific <a> tag within the structure
                profile_link = container.find_element(By.CSS_SELECTOR, "a.EvQUJBaxIRgFetdTQjAXvpGhCNvVbYEbE")
                profile_url = profile_link.get_attribute("href")

                # Filter to include only valid LinkedIn profile links
                if profile_url and "/in/" in profile_url:
                    if profile_url not in unique_urls:
                        unique_urls.add(profile_url)  # Add to the set
                        print(f"Opening profile: {profile_url}")

                        # Open the profile in a new tab
                        driver.execute_script(f"window.open('{profile_url}', '_blank');")
                        time.sleep(3)  # Allow tab to load (increased time)

                        # Switch to the new tab
                        driver.switch_to.window(driver.window_handles[-1])
                        time.sleep(10)  # Increased wait time to ensure the profile page has fully loaded

                        # Scrape the name of the profile
                        try:
                            main_div = driver.find_element(By.CSS_SELECTOR, "div.ph5.pb5")  # Locate the main container
                            name_element = main_div.find_element(By.CSS_SELECTOR, "h1.lSVmRsHWVyNnNULXTOLIQEQVzhkaazrTFzDYEg")  # Locate the name element
                            name = name_element.text
                            print(f"Name scraped: {name}")
                        except Exception as e:
                            print(f"Error scraping name: {e}")

                        # Scrape the headline of the profile
                        try:
                            headline_element = main_div.find_element(By.CSS_SELECTOR, "div.text-body-medium.break-words")  # Locate the headline element
                            headline = headline_element.text
                            print(f"Headline scraped: {headline}")
                        except Exception as e:
                            print(f"Error scraping headline: {e}")

                        # Scrape the most recent experience (Job Title)
                        try:
                            # Scroll to make sure the experience section is loaded
                            scroll_to_load_experience()
                            print("Scrolled Experience Section")

                            # Wait until the experience section is visible and loaded
                            experience_section = WebDriverWait(driver, 10).until(
                                 EC.presence_of_all_elements_located((By.CSS_SELECTOR, "div.display-flex.flex-row.justify-space-between"))
                             )
                            print("Experience section loaded.")
                            if experience_section:
                                print("Experience section Started")
                                # Get the most recent experience (first entry)
                                most_recent_experience = experience_section[0]

                                # Locate the job title inside the most recent experience entry
                                try:
                                    # XPath to select job title in the most recent experience entry
                                    title_element = most_recent_experience.find_element(
                                        By.CSS_SELECTOR,
                                        "div.display-flex.flex-row.justify-space-between > div.display-flex.flex-column.full-width > div.display-flex.flex-wrap.align-items-center.full-height > div.display-flex > div.display-flex.full-width > div.display-flex.align-items-center.mr1.t-bold > span:nth-child(1)"
                                    )
                                    title = title_element.text
                                    print(f"Most Recent Experience Title scraped: {title}")
                                except Exception as e:
                                    print(f"Error scraping most recent experience title: {e}")
                            else:
                                print("No experience section found.")
                        except Exception as e:
                            print(f"Error accessing experience section: {e}")

                        # Increased wait time before closing the profile tab
                        time.sleep(10)  # Longer wait to simulate spending more time on the profile

                        # Close the current tab
                        driver.close()
                        # Switch back to the main tab
                        driver.switch_to.window(driver.window_handles[0])
                        time.sleep(3)  # Allow the main tab to regain focus
                    else:
                        print(f"Skipping duplicate profile: {profile_url}")  # Debugging: Duplicate URL
                else:
                    print(f"Skipping non-profile link: {profile_url}")  # Debugging: Non-profile links
            except Exception as inner_e:
                print(f"Error processing container: {inner_e}")  # Debugging: Container processing error
    except Exception as e:
        print(f"Error while opening profiles: {e}")

# Helper: Navigate to the next page
def go_to_next_page():
    try:
        next_button = WebDriverWait(driver, 10).until(
            EC.element_to_be_clickable((By.CSS_SELECTOR, ".artdeco-pagination__button--next"))
        )
        next_button.click()
        time.sleep(5)  # Allow some time for the next page to load
        return True
    except Exception as e:
        print("Next page not found or end of results:", e)
        return False

# Main logic
def main():
    search_people(SEARCH_QUERY)
    time.sleep(10)

    page = 1

    while page <= 100:  # Limit to 100 pages
        print(f"Processing page {page}...")
        scroll_down()  # Scroll down to ensure all profiles are loaded
        open_profiles()

        # Break if no next page
        if not go_to_next_page():
            break

        page += 1

    print("Profile visiting complete.")

# Run the script
try:
    main()
finally:
    driver.quit()
```
